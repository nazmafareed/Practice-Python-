5. Number of Days Between Two Dates

```python
from datetime import date
date1 = date(2024, 7, 1)
date2 = date(2024, 12, 25)

delta = date2 - date1
print("Number of days between", date1, "and", date2, "is", delta.days)
```

7. Difference Between Given Number and 17

```python
num = int(input("Enter a number: "))

if num > 17:
    print(2 * abs(num - 17))
else:
    print(17 - num)
```

8. Write a Python program to check whether a given number is within 100 of 1000 or 2000.

```python
num = int(input("Enter a number :"))
if abs(num - 1000) <= 100:
    print("True")
elif abs(num - 2000) <= 100:
    print("True")
else:
    print("False")
```

9. Sum of Three Numbers (Triple if All Equal)

```python
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
c = int(input("Enter third number : "))

total = a + b + c 

if a==b==c :
    print(total * 3)
else :
    print(total)
```

10. Check Whether a Letter is a Vowel

```python
letter = input("Enter a letter : ").lower()
if letter in ['a', 'e', 'i', 'o', 'u']:
    print(letter, "is a vowel")
else:
    print(letter, "is a consonant")
```
