Operators in python 

1. **Arithmetic Operators:**  
Arithmetic operators are used to perform mathematical calculations between numbers: 
* ***Addition(+) :*** Adds two operands.
* ***Subtraction(-):*** Subtracts the second operand from the first.
* ***Multiplication(\*):*** Multiplies two operands.
* ***Division(/):*** Divides the first operand by the second.
* ***Modulus(%):*** Returns the remainder after division.
* ***Exponentiation(\*\*):*** Raises the first operand to the power of the second.
* ***Floor Division(//):*** Divides and returns the largest integer (removes decimal part).

```python
# Practical examples of Arithmetic Operators
a = 9
b = 2

#Addition
print("Sum:" ,a+b) #Result: 11

#Subtraction
print(a-b) #Result: 7

#Multiplication
print(a*b) #Result: 18

#Remainder (Modulus operator)
print("Remainder:" ,a % b)  # Output: 1

#power 
print("Square:" ,a ** b) # 9 power 2 is equal to 81(square of 9)

#floor division
print(a // b)  # Output: 4
```

2. **Assingment operators:**  
Assignment operators are used to assign values to variables and perform updates in a single step:
* ***Equal (=):*** Assigns a value to a variable (e.g., x = 5).
* ***Add and Assign (+=):*** Adds a value to the variable and assigns the result (e.g., x += 3 is same as x = x + 3).
* ***Subtract and Assign (-=):*** Subtracts a value and assigns the result (e.g., x -= 2).
* ***Multiply and Assign (\*=):*** Multiplies and assigns the result.
* ***Divide and Assign (/=):*** Divides and assigns the result.

```python
#Practical examples of Assignment Operators
x = 5
y = 10

y += 5 # y = y + 5 
print(y)  # Output: 15

y-=5 # y = y-5
print(y)  #output: 5

x *= 2 # x = x * 2
print(x)
```

3. Comparison (Relational Operators )

```python
a = 10
b = 3 

print( a == b )
print( a != b )
print( a > b )
print(a < b)
print( a >= b )
print(a <= b)
```

```python
4. # Logical Operators and, or, not
# and -> multiplication -> 0 - > false ,1 -> true 
# or -> addition 
# not -> negation

p = 10
q = 4 

print ( (p > 5) and (q < 3) )  # True and false  -> false
print ( (p > 5) or (q < 3) )   # True or false  -> true
print ( not(p > 5) )            # not true -> false
```
